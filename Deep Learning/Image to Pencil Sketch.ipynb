# Importing libraries & data

```python
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("C:\\Users\\Parth Salke\\Downloads\\Datasets\\flower1.jpeg")
plt.imshow(image)
```

# Converion of BGR to RGB Image

```python
rgb_image= cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(rgb_image)
```

# Converion of RGB to Gray Image

```python
gray_image = cv2.cvtColor(rgb_image,cv2.COLOR_RGB2GRAY)
plt.imshow(gray_image)
```

# Inverting the Gray Image

```python
invert_image = cv2.bitwise_not(gray_image)
plt.imshow(invert_image)
```

# Blur the image

```python
blur_image = cv2.GaussianBlur(invert_image,(21,21),sigmaX=0,sigmaY=0)
plt.imshow(blur_image)
```

# Inverting the Blur image

```python
invert_blur = cv2.bitwise_not(blur_image)
plt.imshow(invert_blur)
```

# Making the final pencil sketch

```python
img_divide = cv2.divide(gray_image,invert_blur,scale=260)
pencil_sketch = cv2.cvtColor(img_divide,cv2.COLOR_GRAY2BGR)
plt.imshow(pencil_sketch)
plt.show()
```
