# Salary Vs Age (LR)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']=(20.0,10.0)

data=pd.read_csv('C:\\Users\\Parth Salke\\Downloads\\Datasets\\Salary.csv')
print(data.shape)
data.head(5)
```

```python
data.tail(3)
```

```python
#collecting X and Y
X=data['YearsExperience'].values
Y=data['Salary'].values
```

```python
#Mean X and Y
mean_x=np.mean(X)
mean_y=np.mean(Y)

#total number of values
m=len(X)

#using formula to calculate b1 and b2
numer=0
denom=0
for i in range(m):
    numer +=(X[i]-mean_x)*(Y[i]-mean_y)
    denom +=(X[i]-mean_x)**2
b1=numer/denom
b0=mean_y-(b1*mean_x)
#printing coefficients
print("The value of m:",b1,"and the value of c: ",b0)
```

```python
#Plotting values and regression line
max_x=np.max(X)+100
min_x=np.min(X)-100
#calculate line values of X and Y
x=np.linspace(min_x,max_x,1000)
y=b0+b1*x

#plotting
plt.plot(x,y,color='#58b970',label='Regression Line')
plt.scatter(X,Y,c='#ef5423',label='Scatter Plot')
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("Salary trends")
plt.legend()
plt.show()
```

```python
#find accuracy by R^2 method
ss_t=0
ss_r=0
for i in range(m):
    y_pred=b0+b1*X[i]
    ss_t+=(Y[i]-mean_y)**2
    ss_r+=(Y[i]-y_pred)**2
r2=1-(ss_r/ss_t)
print("Accuracy by R square method: ",r2)
```
